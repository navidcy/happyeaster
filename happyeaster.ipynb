```julia
using PyPlot
```

```julia
n, L = 101, 4.1

x = Array(reshape(range(-L/2, L/2, length=n), (n, 1)))
y = Array(reshape(range(-L/2, L/2, length=n), (1, n)))
X, Y = [ x[i] for i=1:n, j=1:n], [ y[j] for i=1:n, j=1:n]

egg = @. X^2 + Y^2/(1.3 + 0.2Y)^2 - 1

contour(X, Y, egg , levels=(0, 0.01), colors="red")
annotate(L"$x^2 + \frac{y^2}{(1.3+0.2y)^2} = 1$",
    xy=[1; 0],
    xycoords="axes fraction",
    xytext=[-10, 10],
    textcoords="offset points",
    fontsize=16.0,
    ha="right",
    va="bottom")

annotate(L"καλό\ πάσχα",
    xy=[0.5; 1],
    xycoords="axes fraction",
    xytext=[-10, 10],
    textcoords="offset points",
    fontsize=16.0,
    ha="center",
    va="bottom")

xticks(-2:1:2)
yticks(-2:1:2)
axis("square")

savefig("happyeaster.png", dpi=200);
```
